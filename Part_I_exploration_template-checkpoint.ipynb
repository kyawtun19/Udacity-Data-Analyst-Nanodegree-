# Can we calculate the total distance traveled using the start_station_latitude, start_station_longitude, end_station_latitude, and end_station_longitude to determine the most popular station and distances traveled?

```python
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    a = np.sin(delta_phi / 2) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    d = R * c
    return d

# Calculate the distance in kilometers and add it as a new column
df['distance_km'] = df.apply(lambda row: haversine_distance(row['start_station_latitude'], row['start_station_longitude'], row['end_station_latitude'], row['end_station_longitude']), axis=1)

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='distance_km', bins=30)
plt.title('Histogram of Distances Traveled')
plt.xlabel('Distance (km)')
plt.ylabel('Frequency')
plt.show()
```

Most bikers only bike in 0-2 km. Very few bike in between 8-10km. 

```python
start_station_counts = df['start_station_id'].value_counts().reset_index()
start_station_counts.columns = ['start_station_id', 'start_count']

end_station_counts = df['end_station_id'].value_counts().reset_index()
end_station_counts.columns = ['end_station_id', 'end_count']

stations = df[['start_station_id', 'start_station_latitude', 'start_station_longitude']].drop_duplicates()
stations = stations.merge(start_station_counts, on='start_station_id', how='left')
stations = stations.merge(end_station_counts, left_on='start_station_id', right_on='end_station_id', how='left')
stations['total_count'] = stations['start_count'].fillna(0) + stations['end_count'].fillna(0)

stations['normalized_count'] = (stations['total_count'] - stations['total_count'].min()) / (stations['total_count'].max() - stations['total_count'].min())

fig, ax = plt.subplots(figsize=(10, 10))
scatter = ax.scatter(stations['start_station_longitude'], stations['start_station_latitude'], s=stations['normalized_count'] * 500, c=stations['normalized_count'], cmap='YlOrRd', alpha=0.7)

cbar = plt.colorbar(scatter)
cbar.set_label('Popularity')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Popularity of Bike Stations')

ax.set_facecolor('lightblue')
plt.grid(True)
plt.show()

```

It seems that the most popular stations ( in the red ) are located in the latitude of 37.8, and longitude of -122.4. 

Upon researching, they're located in Pier zone, where people can bike with the view of water. Management should put up more bike station in this path.
